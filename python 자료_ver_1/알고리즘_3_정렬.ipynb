## 3-1 버블 정렬 

- 버블 정렬(Bubble sort)은 이웃하는 숫자를 비교하여 작은 수를 앞쪽으로 이동시키는 과정을 반복하여 정렬한다

```python
from IPython.display import Image
Image(filename='버블정렬.png', width= 600, height=500) 
```

```python
list_data = [50,40,90,10]
```

```python
len(list_data)
```

### 버블 정렬(1)
- 밖에 for문은 전체 길이 만큼 순회
- 내부 for문은 두 개의 인덱스 별 순회하면서 비교 

```python
for i in range(1, len(list_data)) :
    for j in range(0, len(list_data)-1) :
        if list_data[j] > list_data[j+1] :
            list_data[j], list_data[j+1] = list_data[j+1], list_data[j]
```

```python
list_data
```

### 버블정렬(2)

- 역순으로 전체 순회 
- 역순으로 하나씩 비교 하면서 순회
- 내부 비교 후 두 위치 변경

```python
list_data = [50,40,90,10]
```

```python
for p in range(len(list_data)-1, 0, -1):
    print("p : ", p)
    for i in range(p):
        if list_data[i] > list_data[i+1]:
            list_data[i], list_data[i+1] = list_data[i+1], list_data[i]
        print(' bubble sort : ', list_data)
                
```

## 3-2 선택 정렬

- 가장 작은 값을 선택해서 맨 앞부터 차례대로 정렬하는 방식
- 알고리즘 문제 풀이에 사용하기는 느린편


```python
from IPython.display import Image
Image(filename='선택정렬.png', width= 200, height=200) 
```

```python
array1 = [7,5,9,0,3,1,6,2,4,8]
```

```python
for i in range(len(array1)) :
    min_index = i # 가장 작은 원소의 인덱스
    ## 가장 작은 값을 찾기
    for j in range(i+1, len(array1)) :
        if array1[min_index] > array1[j] : 
            min_index = j
    # 가장 작은값을 교체하기
    array1[i], array1[min_index] = array1[min_index], array1[i]
    print(" 정렬 변화 ",array1)
```

## 3-4 퀵 정렬
- 퀵 정렬은 불안정 정렬 에 속하며, 다른 원소와의 비교만으로 정렬을 수행하는 비교 정렬 에 속한다.
- 분할 정복 알고리즘의 하나로, 평균적으로 매우 빠른 수행 속도를 자랑하는 정렬 방법
- 합병 정렬(merge sort)과 달리 퀵 정렬은 리스트를 비균등하게 분할한다.
- 분할 정복(divide and conquer) 방법
> - 문제를 작은 2개의 문제로 분리하고 각각을 해결한 다음, 결과를 모아서 원래의 문제를 해결하는 전략이다.
> - 분할 정복 방법은 대개 순환 호출을 이용하여 구현한다.


```python
from IPython.display import Image
Image(filename='퀵정렬.png', width= 400, height=300) 
```

```python
array = [5,3,8,4,9,1,6,2,7]
```

```python
def quick_sort(array, start, end):
    if start >= end: # 원소가 1개인 경우 종료
        return
    pivot = start # 피벗은 첫 번째 원소
    left = start + 1
    right = end
    while(left <= right):
        # 피벗보다 큰 데이터를 찾을 때까지 반복 
        while(left <= end and array[left] <= array[pivot]):
            left += 1
        # 피벗보다 작은 데이터를 찾을 때까지 반복
        while(right > start and array[right] >= array[pivot]):
            right -= 1
        if(left > right): # 엇갈렸다면 작은 데이터와 피벗을 교체
            array[right], array[pivot] = array[pivot], array[right]
        else: # 엇갈리지 않았다면 작은 데이터와 큰 데이터를 교체
            array[left], array[right] = array[right], array[left]
        print('quick sort : ', array)
    # 분할 이후 왼쪽 부분과 오른쪽 부분에서 각각 정렬 수행
    quick_sort(array, start, right - 1)
    quick_sort(array, right + 1, end)

quick_sort(array, 0, len(array) - 1)
print(array)
```

```python
array = [5, 7, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort(array):
    # 리스트가 하나 이하의 원소만을 담고 있다면 종료
    if len(array) <= 1:
        return array

    pivot = array[0] # 피벗은 첫 번째 원소
    tail = array[1:] # 피벗을 제외한 리스트

    left_side = [x for x in tail if x <= pivot] # 분할된 왼쪽 부분
    right_side = [x for x in tail if x > pivot] # 분할된 오른쪽 부분
    print(" sort : ", array)

    # 분할 이후 왼쪽 부분과 오른쪽 부분에서 각각 정렬을 수행하고, 전체 리스트를 반환
    return quick_sort(left_side) + [pivot] + quick_sort(right_side)

print(quick_sort(array))
```

- 정렬한 숫자 개수를 입력한 후에 해당 숫자들을 입력해서 내림차순으로 정렬한다.

### 숫자 개수 입력받기

```python
n = int(input("정렬대상 숫자 입력하기 : "))
```

## 성적 낮은 순으로 출력



```python
n = int(input("정렬대상 숫자 입력하기 : "))
```

```python
array = []

for i in range(n) :
    input_data = input(f" { i+1} 번째 이름과 성적을 입력하세요 :" ).split()
    array.append((input_data[0], int(input_data[1])))
```

```python
array = sorted(array, key=lambda x : x[1])
```
